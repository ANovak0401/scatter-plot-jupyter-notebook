```python
"""create a scatter plot of a random walk"""

import matplotlib.pyplot as plt # import matplotlib and pyplot
from random_walk import RandomWalk # import random_walk.py and class RandomWalk

# make a random walk
rw = RandomWalk() # create an instance of RandomWalk
rw.fill_walk() # fills list of x and y values

#plot the points in the random walk
plt.style.use('dark_background') # set style
fig, ax = plt.subplots() # create plots and figure
ax.scatter(rw.x_values, rw.y_values, s=15) # create scatter chart

plt.show() # render chart window
```

```python
"""styling the chart"""

import matplotlib.pyplot as plt # import matplotlib and pyplot
from random_walk import RandomWalk # import random_walk.py and class RandomWalk

# make a random walk
rw = RandomWalk() # create an instance of RandomWalk
rw.fill_walk() # fills list of x and y values

#plot the points in the random walk
plt.style.use('dark_background') # set style
fig, ax = plt.subplots() # create plots and figure
point_numbers = range(rw.num_points) # generate a list of numbers equal to the points in the walk
ax.scatter(rw.x_values, rw.y_values, c=point_numbers,
           cmap=plt.cm.RdPu, edgecolors='none', s=15) # create scatter chart with x, y values,
                                                      # c= list to control gradient
                                                      # cmap is the colormap to be used
                                                      # edgecolors controls black circles around points
                                                      # s= size of points

plt.show() # render chart window
```

```python
"""plotting the starting and ending points"""

import matplotlib.pyplot as plt # import matplotlib and pyplot
from random_walk import RandomWalk # import random_walk.py and class RandomWalk

# make a random walk
rw = RandomWalk() # create an instance of RandomWalk
rw.fill_walk() # fills list of x and y values

#plot the points in the random walk
plt.style.use('dark_background') # set style
fig, ax = plt.subplots() # create plots and figure
point_numbers = range(rw.num_points) # generate a list of numbers equal to the points in the walk
ax.scatter(rw.x_values, rw.y_values, c=point_numbers,
           cmap=plt.cm.RdPu, edgecolors='none', s=15) # create scatter chart with x, y values,
                                                      # c= list to control gradient
                                                      # cmap is the colormap to be used
                                                      # edgecolors controls black circles around points
                                                      # s= size of points

#emphasize first and last points
ax.scatter(0, 0, c='green', edgecolors='none', s= 100) # marks the starting point with big green dot
ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors= 'none', s= 100) # plot end point red dot

plt.show() # render chart window
```

```python
"""changing to plot"""

import matplotlib.pyplot as plt # import matplotlib and pyplot
from random_walk import RandomWalk # import random_walk.py and class RandomWalk

# make a random walk
rw = RandomWalk() # create an instance of RandomWalk sets number of points to 50000
rw.fill_walk() # fills list of x and y values

#plot the points in the random walk
plt.style.use('dark_background') # set style
fig, ax = plt.subplots(figsize=(15, 9)) # create plots and figure, set figure size

#emphasize first and last points
ax.scatter(0, 0, c='purple', edgecolors='none', s= 100, zorder=10) # marks the starting point with greendot
ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red',
           edgecolors= 'none', s= 100, zorder=5) # plot end point with a red dot
ax.plot(rw.x_values, rw.y_values, linewidth=1, zorder=0) # create scatter chart with x, y values,
                                                         # linewidth = width of line
                                                         # zorder = layer to draw on. higher = deeper

#remove the axes
ax.get_xaxis().set_visible(False) # make x axis invisible
ax.get_yaxis().set_visible(False) # make y axis invisible

plt.show() # render chart window
```
